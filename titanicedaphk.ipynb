```python
#Import statements

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Reading the TITANIC.csv file

teda = pd.read_csv('C:\\Users\\Parjanya\\Downloads\\TITANIC.csv')
```

```python
#Dropping the columns cabin because it has too many NaN values, and others because they are pretty useless

teda_cleaned = teda.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
teda_cleaned.head()

```

```python
#Univariate Analysis
```

```python
#Count plot of the people who survived and those who didn't

sns.set_style('whitegrid')
plt.figure(figsize= (12, 7))
sns.countplot('Survived', data = teda_cleaned)
```

```python
#Count plot of the Pclass

sns.set_style('whitegrid')
plt.figure(figsize= (12, 7))
sns.countplot('Pclass', data = teda_cleaned)
```

```python
#Count plot for sex

sns.set_style('whitegrid')
plt.figure(figsize= (12, 7))
sns.countplot('Sex', data = teda_cleaned)
```

```python
#Count plot for Embarked

sns.set_style('whitegrid')
plt.figure(figsize= (12, 7))
sns.countplot(x = 'Embarked', data = teda_cleaned)
```

```python
#Box plot for Age

sns.set_style('whitegrid')
plt.figure(figsize= (12, 7))
sns.boxplot(y= 'Age', data=teda_cleaned)
```

```python
#Box plot for SibSp

sns.set_style('whitegrid')
plt.figure(figsize= (12, 7))
sns.boxplot(y= 'SibSp', data=teda_cleaned)
```

```python
#distplot for SibSp

sns.set_style('whitegrid')
plt.figure(figsize= (12, 7))
sns.distplot(teda_cleaned['SibSp'], bins=10)
```

```python
#Box plot for Parch

sns.set_style('whitegrid')
plt.figure(figsize= (12, 7))
sns.boxplot(y= 'Parch', data=teda_cleaned)
```

```python
#distplot for Parch

sns.set_style('whitegrid')
plt.figure(figsize= (12, 7))
sns.distplot(teda_cleaned['Parch'], bins=10)
```

```python
#Box plot for Fare

sns.set_style('whitegrid')
plt.figure(figsize= (12, 10))
sns.boxplot(y= 'Fare', data=teda_cleaned)
```

```python
#distplot for Fare

sns.set_style('whitegrid')
plt.figure(figsize= (12, 10))
sns.distplot(teda_cleaned['Fare'], bins=10)
```
